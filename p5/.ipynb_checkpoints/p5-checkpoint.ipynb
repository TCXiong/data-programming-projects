```python
nest ={}

test = data.groupby("ip").size().sort_values()
testdic = test.to_dict()
# for i in testdic:
#     print(i)
    
    
for i in testdic:
    if not ed.lookup_region(i) in nest:
        nest[ed.lookup_region(i)] = {}
        
for i in nest:
    temp7 = dict()
    nest[i] = temp7
    for j in testdic:
        if ed.lookup_region(j) == i:
            if not j in temp7:
                temp7[j] = testdic[j]
            else:
                temp7[j] = temp7[j] + testdic[j]

nest
```

```python
#q11
q11dic = {}
q11test = data.groupby(['cik', 'accession', 'extention']).size().to_dict()
for i in q11test:
    temp = ''
    temp = str(int(i[0]))+ "/" +str(i[1])+ "/" +str(i[2])
    
    for j in fillingObj:
        if temp == j:
            siccode = fillingObj[temp].sic 
            if not siccode is None:
                if not siccode in q11dic:
                    q11dic[siccode] = q11test[i]
                else:
                    q11dic[siccode] = q11test[i] + q11dic[siccode]
                
                
            
q11dic
```

```python
#q12

q12_res = {}
timelist = []
time = data['time'].tolist()
key = 0

for i in range(24):
    timelist.append(i)

for i in time:
#     print(int(i[:2]))
    hour = int(i[:2])
    if not hour in q12_res:
        q12_res[hour] = 1
    else:
        q12_res[hour] = q12_res[hour] + 1
        
q12_res          
```

```python
addr = ["Australia", "France", "Viet Nam"]
q13_dic= {}
data.loc[0,"ip"]
AustraliaState = []
FranceState = []
VietState = []
for i in range(len(data)):
    region = ed.lookup_region(data.loc[i,"ip"])

    if region in addr:
        
        if region == "France":
            html = str(int(data.loc[i,"cik"]))+"/"+str(data.loc[i,"accession"])+"/"+str(data.loc[i,"extention"])
            if html in fillingObj:
                obj = fillingObj[html].state()
                if not obj in FranceState:
                    FranceState.append(obj)
            q13_dic[region] = FranceState
            
        if region == "Australia":
            html = str(int(data.loc[i,"cik"]))+"/"+str(data.loc[i,"accession"])+"/"+str(data.loc[i,"extention"])
        
            if html in fillingObj:
                obj = fillingObj[html].state()
                if not obj in AustraliaState:
                    AustraliaState.append(obj)
            q13_dic[region] = AustraliaState
             
        if region == "Viet Nam":
            html = str(int(data.loc[i,"cik"]))+"/"+str(data.loc[i,"accession"])+"/"+str(data.loc[i,"extention"])
            if html in fillingObj:
                obj = fillingObj[html].state()
                if not obj in VietState:
                    VietState.append(obj)
            q13_dic[region] = VietState
        
q13_dic
```

```python
#q13
from graphviz import Digraph
g = Digraph()

for i in q13_dic:
    g.node(i,i)
    for j in q13_dic[i]:
        g.edge(i,str(j))
    
g
```

```python
#q14
q14_res = {}
tempres = {}
for i in q11test:
    temp = ''
    temp = str(int(i[0]))+ "/" +str(i[1])+ "/" +str(i[2])
    
    for j in fillingObj:
        if temp == j:
            addr = fillingObj[temp].addresses 
            for counter in range(q11test[i]):
                for k in addr:
                    if not k in tempres:
                         tempres[k] = 1
                    else:
                         tempres[k] = tempres[k] + 1
                    
for i in tempres:
    if tempres[i] >= 225:
        q14_res[i] = tempres[i]
        
        
q14_res
```

```python
#q15
df = gpd.read_file("locations.geojson")
shape = gpd.read_file("shapes/cb_2018_us_state_20m.shp")
window = box(-130, 20, -55, 50)

for i in range(len(df['address'])):
    for j in tempres:
        if j==df.loc[i,'address']:
            df.loc[i,"access"] = tempres[j]

ame = shape.intersection(window).to_crs("epsg:3395")
ame = ame[~ame.is_empty]
ax = ame.plot(color="lightgrey", figsize=(4,3))

ame2 = df[df.intersects(window)].to_crs("epsg:3395")
new = ame2.sort_values(by="access")  
ax2 = new.plot(ax = ax , column="access", legend=True, cmap="plasma", figsize=(4,3))


fig = ax2.get_figure()
fig.set_dpi(150)
```
